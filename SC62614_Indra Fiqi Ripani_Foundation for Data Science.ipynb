### **2, 3, 4 Menghitung BMI Setiap Orang dan Menambahnya ke Dalam DataFrame**

2. Hitunglah BMI setiap siswa dengan rumus BMI=Berat badan (kg)(Tinggi Badan (m))2
3. Buatlah program untuk menghitung BMI!
4. Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
# Konversi satuan tinggi badan dari cm ke m
df['Tinggi Badan (m)'] = df['Tinggi Badan (cm)'] / 100

# Menghitung BMI untuk setiap orang
df['BMI'] = df['Berat Badan (kg)'] / (df['Tinggi Badan (m)'] ** 2)
df['BMI'] = df['BMI'].round(3)

# Menampilkan DataFrame dengan kolom BMI
print(df[['No Peserta', 'Nama', 'Berat Badan (kg)', 'Tinggi Badan (m)', 'BMI']].to_string(index=False))
```

### **5 Menentukan Status BMI Setiap Orang**

5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

>  
*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5






```python
# Menentukan status BMI
df['Status BMI'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, float('inf')], labels=['Underweight', 'Normal', 'Overweight', 'Obesity'], right=False)

# Menampilkan DataFrame dengan kolom "Status BMI"
print(df[['No Peserta', 'Nama', 'Berat Badan (kg)', 'Tinggi Badan (m)', 'BMI', 'Status BMI']].to_string(index=False))
```

### **SOAL 2**

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0 - 1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1 - 25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

```python
from babel.numbers import format_currency

def hitung_investasi(modal, tenur):
    persentase_keuntungan = 25

    for tahun in range(1, tenur + 1):
        total_investasi = modal * (1 + persentase_keuntungan / 100) ** tahun
        formatted_investasi = format_currency(total_investasi, 'IDR', locale='id_ID')
        print(f"Tahun {tahun}, total investasi : {formatted_investasi}")

while True:
    try:
        modal = int(input("Modal Investasi (0 - 1.000.000.000) : "))
        if 0 <= modal <= 1000000000:
            break
        else:
            print("Error: Modal investasi harus antara 0 dan 1.000.000.000.")
    except ValueError:
        print("Error: Input harus berupa bilangan bulat.")

while True:
    try:
        tenur = int(input("Jangka Waktu Investasi (1 - 25 tahun) : "))
        if 0 < tenur <= 25:
            break
        else:
            print("Error: Jangka waktu investasi harus antara 1 dan 25 tahun.")
    except ValueError:
        print("Error: Input harus berupa bilangan bulat.")

hitung_investasi(modal, tenur)
```
