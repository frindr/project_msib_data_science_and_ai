## **Versi 1**

### **Import Library Python**

```python
import nltk  # Impor library Natural Language Toolkit (NLTK) untuk pemrosesan bahasa alami.
from sklearn.feature_extraction.text import CountVectorizer  # Impor CountVectorizer dari scikit-learn untuk membuat vektor Bag-of-Words.
from sklearn.naive_bayes import MultinomialNB  # Impor Multinomial Naive Bayes dari scikit-learn untuk klasifikasi teks.
```

### **Memberi Artikel Untuk Training Data**

```python
# Teks/artikel sebagai dasar/pelatihan/training ketiga topik tersebut
text_sport = "Liverpool held off a late charge from Tottenham after Mohamed Salah struck twice to win 2-1 in north London and lift themselves back into the Premier League top-four race. Jurgen Klopp side had suffered shock defeats to relegation-threatened pair Nottingham Forest and Leeds in their last two league outings but started fast against Spurs, with a sharp touch and finish into the bottom corner from in-form Salah giving them the lead on 11 minutes. Ivan Perisic had a header deflected onto the post by Liverpool goalkeeper Alisson and Ryan Sessegnon saw a penalty shout from a challenge by Trent Alexander-Arnold waved away as Spurs came to life - but the Reds struck again just before the break thanks to a gift from Eric Dier. The centre-back miscued a header towards his own goal and Salah (40) raced through to chip in his ninth goal in eight games. Tottenham were sent out early for the second half and - not for the first time this season - were better after the break, with Alisson again pushing a Perisic effort onto the woodwork before Harry Kane (70) fired home a brilliant strike when played in by sub Dejan Kulusevski. Rodrigo Bentancur went close with a couple of headers and Kane glanced wide as Spurs desperately sought another late goal but Liverpool clung on for their first away win in the Premier League this season to move up to eighth and within seven points of fourth-placed Spurs, with a game in hand."
text_medical = "There is growing popularity of the Hospital Incident Command System (HICS) as an organizational tool for hospital management in the COVID-19 pandemic. We specifically describe implementation of HICS at the Isfahan province reference hospital (Isabn-e-Maryam) during the COVID-19 pandemic and try to explore performance of it. Methods: To document the actions taken during the COVID-19 pandemic, standard, open-ended interviews were conducted with individuals occupying activated HICS leadership positions during the event. A checklist based on the job action sheets of the HICS was used for performance assessment. Results: With the onset of the pandemic, hospital director revised ICS structure that adheres to span of better control of COVID-19. Methods of expanding hospital inpatient capacity to enable surge capacity were considered. The highest performance score was in the field of planning. Performance was intermediate in Financial/Administration section and good in other fields. Discussion: In the current COVID-19 pandemic, establishing HICS with some consideration about long-standing events can help improve communication, resource use, staff and patient protection, and maintenance of roles."
text_finance = "According to Fidelity’s Financial Resolutions survey, saving more money was the number one-resolution for respondents. Close to half (43%) said this was a goal they wanted to work toward in the new year. Building a nest egg can help you pay for big ticket items — like a house, a vacation, a wedding or even just an expensive item you really want — without taking on additional debt. Having savings can also come in handy if an emergency expense comes up. If you’re looking to increase your savings in the new year, it helps to start small — even if you’re only transferring $10 a week into your savings account. Starting small helps you build a muscle for saving. This way, when you receive salary bumps, bonuses and gift money, you’ve already gotten into the habit of saving, and you’ll be more likely to transfer that money to your savings account. You may also want to consider automating that process instead of just manually moving money into your savings. Relying on manual transfers leaves a lot of room for procrastination — and before we know it, we’ve spent the money we intended to save. But when you set up automatic transfers into your savings account, you take away the need to make that decision altogether. You can usually schedule automatic transfers through your bank’s mobile app. Lastly, if you want to see your savings grow just a little faster, you can opt for a high-yield savings account instead of a traditional savings account. High-yield savings accounts — like the Marcus by Goldman Sachs Online Savings Account or the Ally Online Savings Account — pay you more in interest each month compared to traditional savings accounts."

# Data latih yang berisi teks-teks dari tiga topik berbeda
training_data = {
    'Kesehatan': [text_medical],
    'Olahraga': [text_sport],
    'Keuangan': [text_finance]
}
```

### **Menghitung Vektor Bag-of-Words Untuk Setiap Topik**

```python
# Membuat objek CountVectorizer untuk mengubah teks menjadi representasi vektor Bag-of-Words
vectorizer = CountVectorizer()

# Inisialisasi list untuk data latih dan label
X_train = []
y_train = []

# Mengisi X_train dan y_train dengan teks dan label dari data latih
for topic, data in training_data.items():
    X_train.extend(data)
    y_train.extend([topic] * len(data))

# Menghitung vektor Bag-of-Words untuk setiap dokumen dalam data latih
X_train_counts = vectorizer.fit_transform(X_train)

# Menampilkan hasil perhitungan
print("Vocabulary:")
print(vectorizer.vocabulary_)  # Menampilkan kamus kata dalam vektor Bag-of-Words
print("\nVektor Bag-of-Words untuk setiap dokumen:")
print(X_train_counts.toarray())  # Menampilkan vektor Bag-of-Words untuk setiap dokumen
```

### **Melatih Model Klasifikasi**

```python
# Membuat objek klasifikasi Naive Bayes Multinomial
classifier = MultinomialNB()

# Melatih model klasifikasi menggunakan vektor Bag-of-Words dan label dari data latih
classifier.fit(X_train_counts, y_train)
```

### **Mengklasifikasikan Suatu Teks**

```python
# Teks yang akan diklasifikasikan
text = "Cristiano Ronaldo came off the bench to earn Manchester United a hard-fought 2-1 victory at Everton in the Premier League on Sunday, taking his career goal tally to 700 in the process. Just as United did last weekend in their derby mauling at the hands of local rivals Manchester City, they again found themselves behind early on at Goodison Park after Alex Iwobi curled a sublime strike into the net from 20 metres."

# Mengubah teks menjadi vektor Bag-of-Words menggunakan objek vectorizer yang sudah dilatih sebelumnya
text_counts = vectorizer.transform([text])

# Memprediksi topik dari teks menggunakan model klasifikasi yang sudah dilatih sebelumnya
predicted_topic = classifier.predict(text_counts)

# Menampilkan topik yang diprediksi
print("Predicted Topic:", predicted_topic[0])
```

### **Import Library Python**

```python
# Mengimpor dan mengunduh paket-paket yang diperlukan dari NLTK
import nltk
nltk.download("popular")

# Mengimpor fungsi-fungsi yang diperlukan dari NLTK
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Mengimpor CountVectorizer dari sklearn untuk membuat vektor Bag-of-Words
from sklearn.feature_extraction.text import CountVectorizer
```

### **Memberikan Artikel Untuk Training Data**

```python
# Teks/artikel sebagai dasar/pelatihan/training ketiga topik tersebut
text_sport = "Liverpool held off a late charge from Tottenham after Mohamed Salah struck twice to win 2-1 in north London and lift themselves back into the Premier League top-four race. Jurgen Klopp side had suffered shock defeats to relegation-threatened pair Nottingham Forest and Leeds in their last two league outings but started fast against Spurs, with a sharp touch and finish into the bottom corner from in-form Salah giving them the lead on 11 minutes. Ivan Perisic had a header deflected onto the post by Liverpool goalkeeper Alisson and Ryan Sessegnon saw a penalty shout from a challenge by Trent Alexander-Arnold waved away as Spurs came to life - but the Reds struck again just before the break thanks to a gift from Eric Dier. The centre-back miscued a header towards his own goal and Salah (40) raced through to chip in his ninth goal in eight games. Tottenham were sent out early for the second half and - not for the first time this season - were better after the break, with Alisson again pushing a Perisic effort onto the woodwork before Harry Kane (70) fired home a brilliant strike when played in by sub Dejan Kulusevski. Rodrigo Bentancur went close with a couple of headers and Kane glanced wide as Spurs desperately sought another late goal but Liverpool clung on for their first away win in the Premier League this season to move up to eighth and within seven points of fourth-placed Spurs, with a game in hand."
text_medical = "There is growing popularity of the Hospital Incident Command System (HICS) as an organizational tool for hospital management in the COVID-19 pandemic. We specifically describe implementation of HICS at the Isfahan province reference hospital (Isabn-e-Maryam) during the COVID-19 pandemic and try to explore performance of it. Methods: To document the actions taken during the COVID-19 pandemic, standard, open-ended interviews were conducted with individuals occupying activated HICS leadership positions during the event. A checklist based on the job action sheets of the HICS was used for performance assessment. Results: With the onset of the pandemic, hospital director revised ICS structure that adheres to span of better control of COVID-19. Methods of expanding hospital inpatient capacity to enable surge capacity were considered. The highest performance score was in the field of planning. Performance was intermediate in Financial/Administration section and good in other fields. Discussion: In the current COVID-19 pandemic, establishing HICS with some consideration about long-standing events can help improve communication, resource use, staff and patient protection, and maintenance of roles."
text_finance = "According to Fidelity’s Financial Resolutions survey, saving more money was the number one-resolution for respondents. Close to half (43%) said this was a goal they wanted to work toward in the new year. Building a nest egg can help you pay for big ticket items — like a house, a vacation, a wedding or even just an expensive item you really want — without taking on additional debt. Having savings can also come in handy if an emergency expense comes up. If you’re looking to increase your savings in the new year, it helps to start small — even if you’re only transferring $10 a week into your savings account. Starting small helps you build a muscle for saving. This way, when you receive salary bumps, bonuses and gift money, you’ve already gotten into the habit of saving, and you’ll be more likely to transfer that money to your savings account. You may also want to consider automating that process instead of just manually moving money into your savings. Relying on manual transfers leaves a lot of room for procrastination — and before we know it, we’ve spent the money we intended to save. But when you set up automatic transfers into your savings account, you take away the need to make that decision altogether. You can usually schedule automatic transfers through your bank’s mobile app. Lastly, if you want to see your savings grow just a little faster, you can opt for a high-yield savings account instead of a traditional savings account. High-yield savings accounts — like the Marcus by Goldman Sachs Online Savings Account or the Ally Online Savings Account — pay you more in interest each month compared to traditional savings accounts."

# Menggabungkan teks-teks ke dalam sebuah list
texts = [text_sport, text_medical, text_finance]
```

### **Preprocessing Teks dan Membuat Vektor Bag-of-Words**

```python
# Impor WordNetLemmatizer dari pustaka NLTK untuk lemmatization
lemmatizer = WordNetLemmatizer()

# Inisialisasi list untuk menyimpan teks yang sudah diproses dan kunci Bag-of-Words
corpus_texts = []
bow_keys = []

# Memproses setiap teks dalam daftar teks
for text in texts:
    # Tokenisasi teks menjadi kata-kata
    words = word_tokenize(text)
    # Melakukan lemmatization pada setiap kata dan menghapus stopwords
    texts = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    # Menambahkan kata-kata yang sudah diproses ke dalam kunci Bag-of-Words
    bow_keys += texts
    # Menggabungkan kata-kata yang sudah diproses kembali menjadi teks
    text = ' '.join(texts)
    # Menambahkan teks yang sudah diproses ke dalam korpus
    corpus_texts.append(text)

# Mengubah list kunci Bag-of-Words menjadi set untuk menghapus duplikat
bow_keys = set(bow_keys)

# Mencetak kunci Bag-of-Words dan teks korpus yang sudah diproses
print(bow_keys)
print(corpus_texts)
```

### **Menambah Teks yang Akan Diklasifikasi**

```python
# Menentukan teks yang akan diklasifikasikan
query_text = "Cristiano Ronaldo came off the bench to earn Manchester United a hard-fought 2-1 victory at Everton in the Premier League on Sunday, taking his career goal tally to 700 in the process. Just as United did last weekend in their derby mauling at the hands of local rivals Manchester City, they again found themselves behind early on at Goodison Park after Alex Iwobi curled a sublime strike into the net from 20 metres."

# Tokenisasi teks query menjadi kata-kata
query_words = word_tokenize(query_text)

# Melakukan lemmatization pada setiap kata dan menghapus stopwords
query_words_clean = [lemmatizer.lemmatize(word) for word in query_words if word not in set(stopwords.words('english'))]

# Menyaring kata-kata query yang ada dalam kunci Bag-of-Words
query_words_corpus = [word for word in query_words_clean if word in set(bow_keys)]

# Menggabungkan kata-kata yang sudah diproses menjadi teks query yang sesuai dengan korpus
query_text_corpus = ' '.join(query_words_corpus)

# Menambahkan teks query yang sudah diproses ke dalam korpus teks
corpus_texts.append(query_text_corpus)
```

### **Membuat Vektor Bag-of-Words Menggunakan CountVectorizer**

```python
# Membuat objek CountVectorizer untuk mengonversi teks menjadi vektor Bag-of-Words
vectorizer = CountVectorizer()

# Mengonversi seluruh teks dalam korpus menjadi vektor Bag-of-Words
bow_vectors = vectorizer.fit_transform(corpus_texts).toarray()
```

### **Normalisasi Vektor Bag-of-Words**

```python
# Normalisasi vektor Bag-of-Words untuk memastikan panjang vektor setiap dokumen sama
bow_texts_norm = []

# Mengiterasi melalui setiap vektor Bag-of-Words
for bow in bow_vectors:
    # Menghitung panjang vektor menggunakan formula Euclidean
    length = (sum(i * i for i in bow)) ** 0.5
    # Normalisasi vektor dengan membagi setiap nilai dengan panjang vektor
    bow_norm = bow / length
    # Menambahkan vektor yang sudah dinormalisasi ke dalam list
    bow_texts_norm.append(bow_norm)
```

### **Mengklasifikasikan Teks**

```python
# Menghitung kemiripan antara teks query dan setiap topik
similarity_vector = []

# Mengambil vektor norm dari teks query
bow_norm_query = bow_texts_norm[-1]  # Teks query berada pada posisi terakhir dalam list

# Mengiterasi melalui vektor norm dari teks latihan (tidak termasuk teks query)
for bow in bow_texts_norm[:3]:
    # Menghitung kemiripan antara teks query dan teks latihan menggunakan dot product
    similarity = sum(i * j for i, j in zip(bow, bow_norm_query))
    similarity_vector.append(similarity)

# Mencari indeks kemiripan tertinggi
id_max_sim = similarity_vector.index(max(similarity_vector))

# Menampilkan hasil klasifikasi berdasarkan kemiripan tertinggi
if id_max_sim == 0:
    print("The query text is classified as: Sport with similarity score:", similarity_vector[0])
elif id_max_sim == 1:
    print("The query text is classified as: Medical with similarity score:", similarity_vector[1])
elif id_max_sim == 2:
    print("The query text is classified as: Finance with similarity score:", similarity_vector[2])
```
